```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Brian\Downloads\CSV_and_Pandas\acc_players-2324F.csv")

col = pd.read_csv(r"C:\Users\Brian\Downloads\CSV_and_Pandas\acc_players-2324F.csv", header=1)
print(df.head)
```

```python
total_points = col['PTS'].sum()
print(f"Total points scored by all players: {total_points}")
```

```python
player_most_minutes = col.loc[col['MP'].idxmax()]
print(f"Player with the most minutes played: {player_most_minutes['Player']}")
```

```python
top_5_rebounds = col.nlargest(5, 'TRB')
print(top_5_rebounds[['Player', 'TRB']])
```

```python
df_filtered = col[col['MP'] > 500]
```

```python
player_most_assists = df_filtered.loc[df_filtered['AST'].idxmax()]
print(f"Player with the highest assists: {player_most_assists['Player']}")
```

```python
top_3_assist_leaders = df_filtered.nlargest(3, 'AST')
print(top_3_assist_leaders[['Player', 'AST']])
```

```python
top_3_blockers = df_filtered.nlargest(3, 'BLK')
print(top_3_blockers[['Player', 'BLK']])
```

```python
school_points = col.groupby('School')['PTS'].sum()
print(school_points)
```

```python
school_assists = col.groupby('School')['AST'].sum()
print(school_assists)
```

```python
top_3_schools_points = school_points.nlargest(3)
print(top_3_schools_points)
```

```python
import matplotlib.pyplot as plt

top_5_players = col.nlargest(5, 'PTS')

plt.bar(top_5_players['Player'], top_5_players['PTS'])
plt.xlabel('Player')
plt.ylabel('Total Points')
plt.title('Top 5 Players by Total Points')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.scatter(col['FG%'], col['WS'])
plt.xlabel('Field Goal Percentage (FG%)')
plt.ylabel('Win Shares (WS)')
plt.title('Correlation Between FG% and Win Shares')
plt.show()

correlation = col['FG%'].corr(col['WS'])
print(f"Correlation coefficient between FG% and WS: {correlation}")
```
